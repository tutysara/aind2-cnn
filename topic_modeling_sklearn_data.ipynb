```python
# read file and save it dict with filename, id and contents
reg = re.compile(r"([^-]+)-([^.]+)")
#dir_name = "/home/tutysara/src/myprojects/ytube/mental_toughness_playlst/txt"
dir_name ="/home/tutysara/src/myprojects/ytube/elliotsaidwhat/txt"
data = []
file_details = []
for file_name in os.listdir(dir_name):
    if file_name == "allcombined.txt":
        continue
    matches = reg.match(file_name)
    title = matches.group(1)
    id = matches.group(2)
    print("processing {title}".format(title=title))
    file = open(dir_name +os.sep+file_name, "r")
    file_data = file.read()
    data.append(file_data)
    file_details.append(file_name)
    
```

```python
#pre process with spacy and replace each word with its lemma
import spacy
nlp = spacy.load('en')
```

```python
new_data = []
for d in data:
    doc = nlp(d)
    new_d = []
    for w in doc:
        if w.is_alpha and not w.is_space and not w.is_stop and w.lemma_ != "-PRON-" :
            #print(w.text, w.lemma_)
            new_d.append(w.text)
            #print("if", w.text, w.is_alpha, not w.is_space, not w.is_stop, w.lemma_ != "-PRON-")
        else:
            pass
            #print("else", w.text, w.is_alpha, not w.is_space, not w.is_stop, w.lemma_ != "-PRON-")
    new_data.append(" ".join(new_d))
        
```
